```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/Predicted_data mlpa.xlsx')
df.head(5)
```

```python
data = df.drop(columns=['s_id','name','profile_link','gender','other_skills'])
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['cgpa'].fillna(data['cgpa'].mean(),inplace=True)
data['inter_gpa'].fillna(data['inter_gpa'].mean(),inplace=True)
data['ssc_gpa'].fillna(data['ssc_gpa'].mean(),inplace=True)
data['internships'].fillna(data['internships'].mean(),inplace=True)
data['no_of_projects'].fillna(data['no_of_projects'].mean(),inplace=True)
data['is_participate_hackathon'].fillna(data['is_participate_hackathon'].mean(),inplace=True)
data['is_participated_extracurricular'].fillna(data['is_participated_extracurricular'].mean(),inplace=True)
data['no_of_programming_languages'].fillna(data['no_of_programming_languages'].mean(),inplace=True)
data['dsa'].fillna(data['dsa'].mean(),inplace=True)
data['mobile_dev'].fillna(data['mobile_dev'].mean(),inplace=True)
data['web_dev'].fillna(data['web_dev'].mean(),inplace=True)
data['Machine Learning'].fillna(data['Machine Learning'].mean(),inplace=True)
data['cloud'].fillna(data['cloud'].mean(),inplace=True)
data['is_placed'].fillna(data['is_placed'].mean(),inplace=True)
data['salary_as_fresher'].fillna(data['salary_as_fresher'].mean(),inplace=True)
data.isnull().sum()
```
